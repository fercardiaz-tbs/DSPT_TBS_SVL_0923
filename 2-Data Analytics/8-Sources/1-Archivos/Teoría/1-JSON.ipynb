#### Obtención de datos

Primero vamos a probar a sacar todos los nombres de los paises

````python
datos = {'names': []}

for i in data:
    nombre_pais = i.get('name', 'NO NAME')
    
    datos['names'].append(nombre_pais)
    #datos['names'] = datos['names'] + [nombre_pais]
    
pd.DataFrame(datos)
````

Ahora obtendremos más info, como el `index` y la url con la imagen del mapa del país, es decir, `map`

````python
datos = {'names': [],
        'index': [],
        'map': []
        }

for i in data:
    nombre_pais = i.get('name', 'NO NAME')
    nombre_index = i.get('index', 'NO INDEX')
    nombre_map = i.get('map', 'NO MAP')
    
    datos['names'].append(nombre_pais)
    datos['index'].append(nombre_index)
    datos['map'].append(nombre_map)
    
pd.DataFrame(datos).head()
````

Pero, ¿y si queremos acceder a datos con un nivel de profundidad mayor? Como por ejemplo la info que hay en `figure`

````python
data[2].get('overview')
````

````python
todos = data
````

````python
datos = {'names': [],
        'index': [],
        'map': [],
         'fig_name': [],
         'overview_date': []
        }

for i in todos:
    nombre_pais = i.get('name', 'NO NAME')
    nombre_index = i.get('index', 'NO INDEX')
    nombre_map = i.get('map', 'NO MAP')
    date_overview = i.get('overview','no overview').get('date', 'no overview date')
    
    # if i.get('figures') != None and type(i.get('figures'))==list():
    for j in i['figures']:
        nombre_fig = j.get('name', 'NO FIGURE NAME')
        
        datos['names'].append(nombre_pais)
        datos['index'].append(nombre_index)
        datos['map'].append(nombre_map)
        datos['fig_name'].append(nombre_fig)
        datos['overview_date'].append(date_overview)
    
df = pd.DataFrame(datos)
df.head()
````
